<ul>Escribe un programa que muestre por consola (con un print) los números de 1 a 100 (ambos incluidos y con un salto de línea entre cada impresion)
Sustituyendo los siguientes:
    <li>Múltiplos de 3 por la palabra "fizz".</li>
    <li>Múltiplos de 5 por la palabra "buzz".</li>
    <li>Múltiplos de 3 y de 5 a la vez por la palabra "fizzbuzz".</li>
</ul>


```python
# Creando la lista de números del 1 al 100 y declarando la nueva lista
lista_numeros = [i for i in range(1,101)]
nueva_lista = []
# Reemplazando valores
for numero in lista_numeros:
    if numero % 5 == 0 and numero % 3 ==0:
        print("fizzbuzz")
    elif numero % 5 == 0:
        print("buzz")
    elif numero % 3 == 0:
        print("fizz")
    else:
        print(numero)
```
